```python
import yfinance as yf
import talib
```

```python
stock = yf.Ticker('PLTR')
hist=stock.history(start="2020-01-01",  end="2021-06-18")
hist
```

```python
from talib import MA_Type

upper, middle, lower = talib.BBANDS(hist.Close, matype=MA_Type.T3)


```

```python
import pandas as pd

tecInd=pd.DataFrame({"date":hist.index,"close":hist.Close,"BBupper":upper,"BBmiddle":middle, "BBlower":lower})
```

```python
import plotly.express as px


fig = px.line(tecInd, x='date', y='close', title='Time Series with Range Slider and Selectors')

for col in ["BBupper","BBmiddle", "BBlower"]:
  fig.add_scatter(x=tecInd['date'], y=tecInd[col],mode='lines')

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig.show()
```
